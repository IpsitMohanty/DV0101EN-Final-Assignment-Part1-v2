````python
import requests
import io
import pandas as pd
import matplotlib.pyplot as plt

# URL of the CSV file
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"

# Fetch the data from the URL
response = requests.get(URL)

# Raise an error if the request failed
response.raise_for_status()

# Convert the response content into a readable format for pandas
csv_content = io.StringIO(response.text)

# Read the CSV data into a pandas dataframe
df = pd.read_csv(csv_content)

# Group the data by 'Year' and calculate the average 'Automobile_Sales'
annual_sales = df.groupby('Year')['Automobile_Sales'].mean()

# List of years from 1980 to 2023
years = list(range(1980, 2024))

# Plot the data with all years on x-axis
plt.figure(figsize=(12, 6))
annual_sales.plot(kind='line', color='b', marker='o', linestyle='-', linewidth=2)

# Set x-ticks for every year between 1980 and 2023
plt.xticks(years, rotation=45)

# Add labels and title
plt.title('Automobile Sales during Recession', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Automobile Sales', fontsize=12)

# Annotating two recession periods (e.g., 1980 and 2007-2009)
plt.text(1980, annual_sales.loc[1980] + 100, '1980 Recession', fontsize=12, color='red')
plt.text(2007, annual_sales.loc[2007] + 100, '2007-2009 Recession', fontsize=12, color='red')

# Display the plot with grid and labels
plt.grid(True)

# Save the plot
plt.savefig('Line_Plot_1.png')

# Show the plot
plt.show()
````

````python
import requests
import io
import pandas as pd
import matplotlib.pyplot as plt

# URL of the CSV file
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"

# Fetch the data from the URL
response = requests.get(URL)

# Raise an error if the request failed
response.raise_for_status()

# Convert the response content into a readable format for pandas
csv_content = io.StringIO(response.text)

# Read the CSV data into a pandas dataframe
df = pd.read_csv(csv_content)

# Filter the dataset for recession periods (where 'Recession' is 1)
recession_df = df[df['Recession'] == 1]

# Group by 'Year' and 'Vehicle_Type', calculate the average 'Automobile_Sales'
recession_sales = recession_df.groupby(['Year', 'Vehicle_Type'], as_index=False)['Automobile_Sales'].mean()

# Pivot the dataframe to have 'Year' as the index and separate lines for each 'Vehicle_Type'
recession_sales_pivot = recession_sales.pivot(index='Year', columns='Vehicle_Type', values='Automobile_Sales')

# Save the plot as "Line_Plot_2.png"
plt.figure(figsize=(12, 6))
recession_sales_pivot.plot(kind='line', marker='o', linewidth=2)

# Add labels and title
plt.title('Automobile Sales Trend by Vehicle Type During Recession Periods', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Automobile Sales', fontsize=12)

# Display the plot with grid and labels
plt.grid(True)
plt.legend(title='Vehicle Type')

# Save the plot
plt.savefig('Line_Plot_2.png')

# Show the plot
plt.show()
````

````python
import requests
import io
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# URL of the CSV file
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"

# Fetch the data from the URL
response = requests.get(URL)

# Raise an error if the request failed
response.raise_for_status()

# Convert the response content into a readable format for pandas
csv_content = io.StringIO(response.text)

# Read the CSV data into a pandas dataframe
df = pd.read_csv(csv_content)

# Group the data by 'Recession' and 'Vehicle_Type', calculate the average 'Automobile_Sales'
avg_sales_by_recession = df.groupby(['Recession', 'Vehicle_Type'], as_index=False)['Automobile_Sales'].mean()

# Plot the data using a bar plot to compare sales during recession and non-recession periods for each vehicle type
plt.figure(figsize=(12, 6))
sns.barplot(x='Vehicle_Type', y='Automobile_Sales', hue='Recession', data=avg_sales_by_recession, palette='coolwarm')

# Add labels and title
plt.title('Comparison of Automobile Sales During Recession and Non-Recession Periods', fontsize=16)
plt.xlabel('Vehicle Type', fontsize=12)
plt.ylabel('Average Automobile Sales', fontsize=12)

# Display the plot with grid
plt.grid(True)
plt.show()
````

### TASK 1.4: Use sub plotting to compare the variations in GDP during recession and non-recession period by developing line plots for each period.
<br>Now, you want to find more insights from the data to understand the reason. <br>Plot a two line charts using subplotting to answer:-
#### How did the GDP vary over time during recession and non-recession periods? 
<br>Make use of <u>add_subplot()</u> from Matplotlib for this comparision.


````python
import requests
import io
import pandas as pd
import matplotlib.pyplot as plt

# URL of the CSV file
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"

# Fetch the data from the URL
response = requests.get(URL)

# Raise an error if the request failed
response.raise_for_status()

# Convert the response content into a readable format for pandas
csv_content = io.StringIO(response.text)

# Read the CSV data into a pandas dataframe
df = pd.read_csv(csv_content)

# Separate the data for recession and non-recession periods
recession_df = df[df['Recession'] == 1]
non_recession_df = df[df['Recession'] == 0]

# Create the subplots for GDP comparison
fig = plt.figure(figsize=(12, 6))

# Add the first subplot for GDP during recession periods
ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(recession_df['Year'], recession_df['GDP'], color='r', marker='o', linestyle='-', linewidth=2)
ax1.set_title('GDP During Recession Periods', fontsize=14)
ax1.set_xlabel('Year', fontsize=12)
ax1.set_ylabel('GDP', fontsize=12)
ax1.grid(True)

# Add the second subplot for GDP during non-recession periods
ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(non_recession_df['Year'], non_recession_df['GDP'], color='g', marker='o', linestyle='-', linewidth=2)
ax2.set_title('GDP During Non-Recession Periods', fontsize=14)
ax2.set_xlabel('Year', fontsize=12)
ax2.set_ylabel('GDP', fontsize=12)
ax2.grid(True)

# Save the plot as "Subplot.png"
fig.savefig('Subplot.png')

# Display the plot
plt.tight_layout()
plt.show()
````

````python
import folium
import pandas as pd

# Sample data with cities, sales, and city locations
# Replace this with your actual data for cities, sales, and location coordinates
sales_data = {
    'City': ['Los Angeles', 'New York', 'Chicago', 'San Francisco', 'Houston'],
    'Sales': [10000, 15000, 12000, 8000, 5000],
    'Latitude': [34.0522, 40.7128, 41.8781, 37.7749, 29.7604],
    'Longitude': [-118.2437, -74.0060, -87.6298, -122.4194, -95.3698],
}

# Create a DataFrame with the sales data
sales_df = pd.DataFrame(sales_data)

# Create a map centered around the US
m = folium.Map(location=[37.0902, -95.7129], zoom_start=5)

# Adding a choropleth map (for demonstration, we'll simulate sales per city)
# Here, the sales column will be used to generate the choropleth colors
folium.Choropleth(
    geo_data='https://raw.githubusercontent.com/PublicaMundi/MappingAPI/master/data/geojson/us-states.json',  # GeoJSON data for the US states
    name='choropleth',
    data=sales_df,
    columns=['City', 'Sales'],
    key_on='feature.properties.name',  # Make sure the GeoJSON's property matches the city name
    fill_color='YlOrRd',  # Color palette
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Sales by City',
).add_to(m)

# Add city markers for the highest sales locations (showing sales in a popup)
for _, row in sales_df.iterrows():
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=10,
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.6,
        popup=f"City: {row['City']}<br>Sales: {row['Sales']}",
    ).add_to(m)

# Save the map as an HTML file
m.save('recession_sales_choropleth.html')

# Display the map
m
````
